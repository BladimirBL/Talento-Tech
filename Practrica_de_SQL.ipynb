##2.1.1 Ejemplo: Listar los países

```python
query = '''SELECT * FROM Country;'''

countries = pd.read_sql(query, conn)
countries
```

```python
query = 'SELECT * FROM \'Country\';' # sin 3 comillas y en cambio usando \

countries = pd.read_sql(query, conn)
countries
```
